```python
import chromadb
from sentence_transformers import SentenceTransformer
```

```python
CHROMA_PATH="/content/chroma_db1"
client = chromadb.PersistentClient(path=CHROMA_PATH)
```

```python
collection = client.get_or_create_collection(name="my_documents")
documents = ["I love machine learning ",
             " The weather is sunny today",
             " Football is a great sport",
             "Natural Language Processing is a part of AI",
             "Deep Learning enables many AI applications"]
```

```python
ids =[f"doc{i}" for i in range(len(documents))]
ids
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(documents).tolist()
embeddings[0]
```

```python
collection.add(documents=documents,embeddings=embeddings,ids=ids)
print("ChromaDB created at "+CHROMA_PATH)
```

```python
query = "AI applications?"
query_embeddings = model.encode([query]).tolist()
```

```python
results =collection.query(query_embeddings=query_embeddings,n_results=2)
#results
```

```python
print("doc: ", results['documents'][0])
print("dist: ", results['distances'][0])
```
